```python
from collections import deque

def Solution(a, b, target):
    # Dictionary to store visited states
    m = {}
    isSolvable = False
    path = []

    # Queue for BFS
    q = deque()

    # Initialize with both jugs empty
    q.append((0, 0))

    while len(q) > 0:
        # Get the current state
        u = q.popleft()

        # If we have already visited this state, continue
        if (u[0], u[1]) in m:
            continue
        
        # If values are out of bounds, continue
        if u[0] > a or u[1] > b or u[0] < 0 or u[1] < 0:
            continue

        # Append the state to the path
        path.append((u[0], u[1]))

        # Mark state as visited
        m[(u[0], u[1])] = 1

        # If we reach the target in either jug
        if u[0] == target or u[1] == target:
            isSolvable = True

            # If result is found in Jug1, empty Jug2 (if needed)
            if u[0] == target and u[1] != 0:
                path.append((u[0], 0))

            # If result is found in Jug2, empty Jug1 (if needed)
            elif u[1] == target and u[0] != 0:
                path.append((0, u[1]))

            # Print the solution path
            print("Path from initial state to solution state:")
            for p in path:
                print(f"({p[0]}, {p[1]})")
            return  # Exit after finding solution

        # Fill Jug2
        q.append((u[0], b))

        # Fill Jug1
        q.append((a, u[1]))

        # Transfer water from Jug1 to Jug2
        for ap in range(max(a, b) + 1):
            c = u[0] + ap
            d = u[1] - ap

            if c == a or (d == 0 and d >= 0):
                q.append((c, d))

            c = u[0] - ap
            d = u[1] + ap

            if (c == 0 and c >= 0) or d == b:
                q.append((c, d))

        # Empty Jug1
        q.append((a, 0))

        # Empty Jug2
        q.append((0, b))

    if not isSolvable:
        print("Solution not possible")

# Driver code
if __name__ == '__main__':
    Jug1, Jug2, target = 4, 3, 2
    Solution(Jug1, Jug2, target)
```
